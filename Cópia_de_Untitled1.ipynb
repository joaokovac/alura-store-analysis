```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando dados das lojas

loja1 = pd.DataFrame({
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor"],
    "Preço": [3500, 120, 200, 900],
    "Avaliação": [4.5, 4.2, 4.1, 4.6],
    "Frete": [40, 15, 20, 35]
})

loja2 = pd.DataFrame({
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor"],
    "Preço": [3300, 110, 190, 850],
    "Avaliação": [4.4, 4.0, 4.2, 4.5],
    "Frete": [45, 18, 22, 30]
})

loja3 = pd.DataFrame({
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor"],
    "Preço": [3000, 100, 180, 800],
    "Avaliação": [4.0, 3.9, 4.1, 4.2],
    "Frete": [50, 20, 25, 40]
})

loja4 = pd.DataFrame({
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor"],
    "Preço": [2800, 90, 170, 750],
    "Avaliação": [3.8, 3.7, 3.9, 4.0],
    "Frete": [55, 25, 30, 45]
})

# Dicionário com todas as lojas

lojas = {
    "Loja 1": loja1,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}

# ---------------------------
# Faturamento por loja
# ---------------------------

faturamento = {}

for nome, loja in lojas.items():
    faturamento[nome] = loja["Preço"].sum()

print("Faturamento por loja")
print(faturamento)

# ---------------------------
# Avaliação média
# ---------------------------

avaliacoes = {}

for nome, loja in lojas.items():
    avaliacoes[nome] = loja["Avaliação"].mean()

print("\nAvaliação média das lojas")
print(avaliacoes)

# ---------------------------
# Frete médio
# ---------------------------

frete = {}

for nome, loja in lojas.items():
    frete[nome] = loja["Frete"].mean()

print("\nFrete médio por loja")
print(frete)

# ---------------------------
# Gráfico faturamento
# ---------------------------

plt.bar(faturamento.keys(), faturamento.values())
plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento")
plt.show()

# ---------------------------
# Gráfico avaliação
# ---------------------------

plt.bar(avaliacoes.keys(), avaliacoes.values())
plt.title("Avaliação Média das Lojas")
plt.xlabel("Lojas")
plt.ylabel("Avaliação")
plt.show()

# ---------------------------
# Gráfico frete
# ---------------------------

plt.pie(frete.values(), labels=frete.keys(), autopct='%1.1f%%')
plt.title("Distribuição do Frete Médio")
plt.show()
```
